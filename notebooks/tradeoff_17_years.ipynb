```python
import plotly.graph_objects as go
import pandas as pd

df_ranked = pd.read_csv('../analysis/top_ranked_17_year.csv')
df_pareto = df_ranked.loc[df_ranked['rank'] == 0]
df_pareto = df_pareto.sort_values(by = 'drawdown_integral')



fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_pareto['drawdown_integral'],
    y=df_pareto['annualized_return'],
    mode='lines+markers',
    name='Pareto Path',
    marker=dict(
        size=10,
        colorscale='Viridis',
        showscale=True,
    ),
    line=dict(width=2, color='rgba(100, 100, 100, 0.5)'),
    # Hover configuration
    customdata=df_pareto[['drawdown_integral', 'annualized_return', 'worst_annual_4wk']],
    hovertemplate=(
        "<b>drawdown:</b> %{customdata[0]}<br>" +
        "<b>return:</b> %{customdata[1]:.4f}<br>" +
        "<b>worst_annual:</b> %{customdata[2]:.4f}<br>" +
        "<extra></extra>" # Removes the secondary trace name box
    )
))

fig.update_layout(
    title="Pareto Path: Crash vs Boom Objectives",
    xaxis_title="Crash Performance",
    yaxis_title="Boom Performance",
    hovermode="closest",
    template="plotly_white",
    width=900,
    height=600
)

fig.show()
```

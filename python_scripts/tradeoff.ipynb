```python
import pandas as pd

df = pd.read_csv('../analysis/top_ranked.csv')
df = df.loc[df['rank'] == 0]
df.loc['drawdown'] = df.loc[:, ['f1_2008', 'f2_2020', 'f3_2022']].sum(axis = 1)
```

```python
import plotly.express as px

# Assuming your dataframe is named 'df'
fig = px.scatter(
    df, 
    x="drawdown", 
    y="f4_terminal",
    hover_data=["f5_annual_worst"],
    title="Portfolio Optimization: Terminal Value vs. Drawdown",
    labels={
        "drawdown": "Drawdown (Cumulative/Max)",
        "f4_terminal": "Terminal Portfolio Value",
        "f5_annual_worst": "Worst Annual Return"
    },
    template="plotly_white"
)

# Improves the hover experience by showing the label clearly
fig.update_traces(marker=dict(size=10, opacity=0.7, line=dict(width=1, color='DarkSlateGrey')))

fig.show()
```
